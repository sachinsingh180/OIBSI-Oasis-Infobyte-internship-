```python
import pandas as pd

# File name(e.g., 'retail_sales_dataset.csv')
df = pd.read_csv('retail_sales_dataset.csv')

# showing data
print("Data ki pehli 5 lines:")
print(df.head())

# check for data basic info
print("\nData Information:")
print(df.info())
# 1. Check any column has any type of missing values
print("Missing values in each column:")
print(df.isnull().sum())

# 2. Check for duplicates rows
print("\nDuplicate rows count:", df.duplicated().sum())

# 3. finding Descriptive Statistics (Mean, Median, Mode)
print("\nBasic Statistics:")
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. show the sales based on product category
plt.figure(figsize=(10, 6))
sns.countplot(x='Product Category', data=df)
plt.title('Sales Count by Product Category')
plt.show()

# 2. Gender based sales report
plt.figure(figsize=(8, 5))
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Sales Distribution by Gender')
plt.show()
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Ye check the product sells more quantity
plt.figure(figsize=(10, 6))
sns.countplot(x='Product Category', data=df, palette='viridis')
plt.title('Sales Count by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales Count')
plt.show()

# 2. check for male or female customer ratio
plt.figure(figsize=(8, 8))
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
plt.title('Sales Distribution by Gender')
plt.axis('equal')
plt.show()
```

```python
# 1. to convert the correct formet
df['Date'] = pd.to_datetime(df['Date'])

# 2. count total sales based on per months
df['Month'] = df['Date'].dt.strftime('%Y-%m')
monthly_sales = df.groupby('Month')['Total Amount'].sum()

# 3. Line Graph
plt.figure(figsize=(12, 6))
monthly_sales.plot(kind='line', marker='o', color='darkorange', linewidth=2)
plt.title('Monthly Sales Trend (Time Series)')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
