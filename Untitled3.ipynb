```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Dataset load karein
df = pd.read_csv('AB_NYC_2019.csv')

# 2. Check karein kitni missing values hain
print("Missing values before cleaning:")
print(df.isnull().sum())

# 3. Data Cleaning
# 'name' aur 'host_name' ki missing rows ko hata dete hain
df.dropna(subset=['name', 'host_name'], inplace=True)

# 'last_review' aur 'reviews_per_month' mein null values ko 0 ya 'Unknown' se bhar dete hain
df['reviews_per_month'].fillna(0, inplace=True)
df['last_review'].fillna('No Review', inplace=True)

# 4. Check karein ab koi null value bachi hai?
print("\nMissing values after cleaning:")
print(df.isnull().sum())

# 5. Ek chota sa Visualization (Room Type Distribution)
plt.figure(figsize=(10,6))
sns.countplot(x='room_type', data=df, palette='viridis')
plt.title('Distribution of Room Types in NYC')
plt.show()
```
